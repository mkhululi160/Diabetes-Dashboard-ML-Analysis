```python
# invalid values
columns = [
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI",
]

(df[columns] == 0).sum()
```

```python
# replace invalid missing values
df[columns] = df[columns].replace(0, np.nan)
```

```python
df.isnull().sum()
```

```python
# fill missing values
for column in columns:

    df[column].fillna(df[column].median(),
                      inplace=True)
```
